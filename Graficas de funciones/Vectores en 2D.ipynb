```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
# vectores = quiver

#Datos
x = np.linspace(-2,2,20)
y = np.linspace(-2,2,20)

xmesh,ymesh = np.meshgrid(x,y)

def V(x,y):
    p = x*y
    q = y**2 - x**2
    return np.array([p,q])

pmesh, qmesh = V(xmesh,ymesh)

#plt.scatter(xmesh,ymesh)
plt.quiver(xmesh,ymesh,pmesh,qmesh)
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(6,6))
axes.axis([-2,2,-2,2])
axes.axvline(x=0,ls='--',lw=2,color='k')
axes.axhline(y=0,ls='--',lw=2,color='k')
axes.quiver(xmesh,ymesh,pmesh,qmesh,color='red')
```
