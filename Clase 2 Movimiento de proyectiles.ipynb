```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import constants as const
```

```python
def Y(t,v0):
    return v0*t - (1/2)*const.g*(t**2)

def Vy(t,v0):
    return v0 - const.g*t
```

```python
v0 = (32 + (9*const.g))/12 
T = np.linspace(0,2,100)
y = [Y(t,v0) for t in T]
v = [Vy(t,v0) for t in T]
```

```python
fig, axes = plt.subplots(figsize=(6,6))
axes.axis([-0.1,2,-5,6])
axes.axvline(x=0,ls='--',lw=2,color='k')
axes.axhline(y=0,ls='--',lw=2,color='k')
axes.plot(T,y)
axes.plot(1.5,Y(1.5,v0),marker='o',color='red')
axes.vlines(1.5, 0, 6,ls='--',lw=1,color='black')
axes.hlines(4, 0, 2,ls='--',lw=1,color='black')
axes.set_title("Cambio de la posición con respecto al tiempo")
axes.set_xlabel("Tiempo (s)")
axes.set_ylabel("Posición vertical (m)")
#Velocidad
axes.plot(T,v,color='pink')
```

```python
d=6
h=5
s=1.8
v0=30
theta0 = np.arctan((h-s)/d)
thau = d/(v0*np.cos(theta0))
#Apple:
def Xa(t):
    return d
def Ya(t):
    return h - (1/2)*const.g*(t**2)
#Bullet:
def Xb(t):
    return v0*t*np.cos(theta0)
def Yb(t):
    return s + v0*t*np.sin(theta0) - (1/2)*const.g*(t**2)
```

```python
delta = 10
T = np.linspace(0,int(theta0)+delta,1000)
xb = [Xb(t) for t in T]
yb = [Yb(t) for t in T]
xa = [Xa(t) for t in T]
ya = [Ya(t) for t in T]
```

```python
fig, axes = plt.subplots(figsize=(6,6))
axes.axis([-1,10,-1,11])
axes.axvline(x=0,ls='--',lw=2,color='k')
axes.axhline(y=0,ls='--',lw=2,color='k')
axes.hlines(h, 0, 10,ls='--',lw=1,color='black')
axes.hlines(s, 0, 10,ls='--',lw=1,color='black')
axes.plot(Xb(thau),Yb(thau),marker='o',color='red')
axes.plot(xb,yb)
axes.plot(xa,ya)
```
